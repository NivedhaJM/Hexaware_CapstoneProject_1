```python
# Data generation for orders.csv

import csv
import random
from datetime import datetime, timedelta
import os

# Configuration
BASE_DIR = "/Volumes/dev/retail_db/dataset/data"
OUTPUT_DIR = os.path.join(BASE_DIR, "orders")
OUTPUT_FILE = "orders.csv"
NUM_RECORDS = 8000

os.makedirs(OUTPUT_DIR, exist_ok=True)
file_path = os.path.join(OUTPUT_DIR, OUTPUT_FILE)

store_codes = ["CHN001", "BLR001", "HYD001", "MUM001", "DEL001", "KOL001", "PUN001", "AHM001", "JAI001", "LKO001",
    "CBE001", "VIZ001", "KOC001", "NAG001", "GOA001"]

start_date = datetime(2024, 1, 1)
end_date = datetime(2026, 1, 1)

def random_date(start, end):
    delta = end - start
    return start + timedelta(days=random.randint(0, delta.days))

with open(file_path, mode="w", newline="", encoding="utf-8") as file:

    writer = csv.writer(file)

    writer.writerow([
        "OrderID",
        "CustomerID",
        "ProductID",
        "OrderDate",
        "Quantity",
        "UnitPrice",
        "StoreCode"
    ])

    for oid in range(1, NUM_RECORDS ):
        customer_id = random.randint(5000, 8500)

        # 95% valid ProductIDs, 5% invalid
        if random.random() < 0.95:
            product_id = f"P{random.randint(101, 999)}"
        else:
            product_id = f"P{random.randint(1000, 1700)}"

        quantity = random.choice([
            random.randint(1, 25),0,-1
        ])

        unit_price = random.choice([
            round(random.uniform(100, 5000), 2),
            "",
            None
        ])

        writer.writerow([
            oid,
            customer_id,
            product_id,
            random_date(start_date, end_date).strftime("%Y-%m-%d"),
            quantity,
            unit_price,
            random.choice(store_codes)
        ])

print(f"Generated {NUM_RECORDS} records at:")
print(file_path)
```

```python
# Data generation for products.csv

import csv
import random
import os

# Configuration

BASE_DIR = "/Volumes/dev/retail_db/dataset/data"
OUTPUT_DIR = os.path.join(BASE_DIR, "products")
OUTPUT_FILE = "products.csv"
NUM_RECORDS = 900


os.makedirs(OUTPUT_DIR, exist_ok=True)
file_path = os.path.join(OUTPUT_DIR, OUTPUT_FILE)

products = {

        "Electronics": {
            "Smartphones": ["Apple", "Samsung", "OnePlus", "Google", "Xiaomi", "Nothing"],
            "Laptops": [ "Dell", "HP", "Lenovo", "Asus", "Acer", "MSI"],
            "Televisions": [ "Sony", "LG", "Samsung", "Panasonic", "TCL"]
        },

        "Fashion": {
            "Men Clothing": [ "Nike", "Adidas", "Puma", "Levi's", "Under Armour" ],
            "Women Clothing": [ "Zara", "H&M", "Forever 21", "Uniqlo", "Calvin Klein"]
        },

        "Sports": {
            "Fitness": [ "Nike", "Adidas", "Reebok", "Puma", "Decathlon"],
            "Cricket": [ "MRF", "SG", "SS", "Kookaburra", "Gray-Nicolls"]
        },

        "Home Appliances": {
            "Refrigerators": [ "LG", "Samsung", "Whirlpool", "Bosch", "Haier", "Godrej" ],
            "Washing Machines": [ "IFB", "Bosch", "LG", "Samsung", "Whirlpool", "Panasonic"]
        },
    
}

with open(file_path, mode="w", newline="", encoding="utf-8") as file:

    writer = csv.writer(file)

    writer.writerow([
        "ProductID",
        "ProductName",
        "Category",
        "SubCategory",
        "Brand",
        "CostPrice"
    ])

    for pid in range(101, 101 + NUM_RECORDS):

        category = random.choice(list(products.keys()))
        sub_category = random.choice(list(products[category].keys()))
        brand = random.choice(products[category][sub_category])

        product_name = f"{brand} {sub_category} Model-{random.randint(100,999)}"

        cost_price = random.choice([
            round(random.uniform(100, 5000), 2),
            "",
            None
        ])

        writer.writerow([
            f"P{pid}",
            product_name,
            category,
            sub_category,
            brand,
            cost_price
        ])

print(f"Generated {NUM_RECORDS} records at:")
print(file_path)
```

```python
# Data generation for customers.csv

import csv
import random
from datetime import datetime, timedelta
import os

# Configuration
BASE_DIR = "/Volumes/dev/retail_db/dataset/data"
OUTPUT_DIR = os.path.join(BASE_DIR, "customers")
OUTPUT_FILE = "customers.csv"
NUM_RECORDS = 3500

os.makedirs(OUTPUT_DIR, exist_ok=True)
file_path = os.path.join(OUTPUT_DIR, OUTPUT_FILE)

first_names = ["John", "Alice", "Bob", "Emma", "David", "Sophia", "James", "Olivia", "Michael", "Emily", "Daniel", "Ava", "Matthew", "Mia", "Ethan", "Charlotte", "William", "Amelia", "Alexander", "Harper", "Benjamin"]
last_names = ["Smith", "Johnson", "Brown", "Davis", "Wilson", "Taylor", "Parker", "Collins", "Edwards", "Stewart",
"Morris", "Murphy", "Cook", "Rogers", "Morgan", "Peterson", "Cooper"]

cities_states = [
    ("Chennai", "Tamil Nadu"),
    ("Bangalore", "Karnataka"),
    ("Hyderabad", "Telangana"),
    ("Mumbai", "Maharashtra"),
    ("New Delhi", "Delhi"),
    ("Kolkata", "West Bengal"),
    ("Pune", "Maharashtra"),
    ("Ahmedabad", "Gujarat"),
    ("Jaipur", "Rajasthan"),
    ("Lucknow", "Uttar Pradesh"),
    ("Kochi", "Kerala"),
    ("Coimbatore", "Tamil Nadu")
]

start_date = datetime(2024, 1, 1)

with open(file_path, mode="w", newline="", encoding="utf-8") as file:

    writer = csv.writer(file)

    writer.writerow([
        "CustomerID",
        "FirstName",
        "LastName",
        "Email",
        "Phone",
        "City",
        "State",
        "LastUpdated"
    ])

    for cid in range(3501, 3500 + NUM_RECORDS):

        first_name = random.choice(first_names)
        last_name = random.choice(last_names)

        city, state = random.choice(cities_states)

        if random.random() < 0.95:
            email = f"{first_name.lower()}.{last_name.lower()}{cid}@gmail.com"
        else:
            email = ""

        phone = random.choice([
            f"9{random.randint(100000000,999999999)}",
            ""
        ])

        last_updated = (
            start_date +
            timedelta(days=random.randint(0, 730))
        ).strftime("%Y-%m-%d %H:%M:%S")

        writer.writerow([
            cid,
            first_name,
            last_name,
            email,
            phone,
            city,
            state,
            last_updated
        ])

print(f"Generated {NUM_RECORDS} records at:")
print(file_path)
```

```python
import json
import random
from datetime import datetime, timedelta
import os

# Configuration
BASE_DIR = "/Volumes/dev/retail_db/dataset/data"
OUTPUT_DIR = os.path.join(BASE_DIR, "exchange_rates")
OUTPUT_FILE = "exchange_rates.json"

os.makedirs(OUTPUT_DIR, exist_ok=True)
file_path = os.path.join(OUTPUT_DIR, OUTPUT_FILE)

currencies = [
    "EUR", "GBP", "INR", "JPY", "AUD",
    "CAD", "CHF", "SGD", "CNY", "NZD", "AED"
]

start_date = datetime(2025, 1, 1)

payload = {"exchange_rates": []}

for i in range(70):

    current_date = (
        start_date + timedelta(days=i)
    ).strftime("%Y-%m-%d")

    rates = {}

    for currency in currencies:
        rates[currency] = round(
            random.uniform(0.5, 150),
            6
        )

    payload["exchange_rates"].append({
        "base_currency": "USD",
        "rate_date": current_date,
        "rates": rates
    })

with open(file_path, "w") as file:
    json.dump(payload, file, indent=4)

print(f"Generated exchange_rates.json at:\n{file_path}")
```
